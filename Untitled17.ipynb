```python
N = 8

def print_board(board):
    for row in board:
        print(" ".join("Q" if col else "." for col in row))
    print()

def is_safe(board, row, col):
    # Check left side of the row
    for i in range(col):
        if board[row][i]:
            return False

    # Check upper-left diagonal
    i, j = row, col
    while i >= 0 and j >= 0:
        if board[i][j]:
            return False
        i -= 1
        j -= 1

    # Check lower-left diagonal
    i, j = row, col
    while i < N and j >= 0:
        if board[i][j]:
            return False
        i += 1
        j -= 1

    return True

def solve(board, col):
    # Base case: all queens are placed
    if col == N:
        print_board(board)
        return True

    res = False
    for row in range(N):
        if is_safe(board, row, col):
            board[row][col] = 1
            res = solve(board, col + 1) or res
            board[row][col] = 0  # Backtrack

    return res

# Initialize the board
board = [[0 for _ in range(N)] for _ in range(N)]

# Solve the problem
solve(board, 0)
```
